```python
'''
3. Pesquise o algoritmo do "Crivo de Eratóstenes" - um dos primeiros
métodos para encontrar números primos, concebido séculos antes da invenção do computador.
Implemente um programa que imprima os números primos entre 0 e 100.
'''
```

```python
def crivo_de_eratostenes(n):
    primo = [True for _ in range(n+1)]
    p = 2
    while p**2 <= n:
        if primo[p]:
            for i in range(p**2, n+1, p):
                primo[i] = False
        p += 1
    for p in range(2, n+1):
        if primo[p]:
            print(p, end=" ")

print("Números primos entre 0 e 100:")
crivo_de_eratostenes(100)
```
