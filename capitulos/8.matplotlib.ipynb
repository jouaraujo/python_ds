```python
plt.plot([-2, -1, 0, 1, 2]);
```

```python
plt.plot([-2, -1, 0, 1, 2])
plt.ylabel('Legenda do Eixo Y');
```

```python
plt.plot([-2, -1, 0, 1, 2])
plt.ylabel('Legenda do Eixo Y')
plt.xlabel('Legenda do Eixo X');
```

```python
plt.plot([-2, -1, 0, 1, 2, 3], [4, 1, 0, 1, 4, 9])
plt.xlabel('Legenda do Eixo X')
plt.ylabel('Legenda do Eixo Y');
```

```python
plt.plot([-2, -1, 0, 1, 2, 3], [4, 1, 0, 1, 4, 9], 'D--r')
plt.xlabel('Legenda do Eixo X')
plt.ylabel('Legenda do Eixo Y');
```

```python
from random import randint
```

```python
cores = []
for k in range(0, 30):
    cores.append(randint(0, 90))

cores
```

```python
# gráficos de dispersão
```

```python
import pandas as pd
```

```python
dataframe = pd.read_csv('../dados/tempo_salarios.csv')
```

```python
plt.scatter(dataframe['Salario'], dataframe['AnosdeExperiencia'], c=cores, s=100) 
plt.xlabel('Salário')
plt.ylabel('Anos de experiência')
plt.show()
```
